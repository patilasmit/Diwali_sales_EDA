Explotary data analysis

1. Which are most buying in with high purchasing power between male and female?

```python
sales.columns
```

```python
ax = sns.countplot(x ='Gender', data=sales)
for bars in ax.containers:
    ax.bar_label(bars)
```

Sales :

```python
sales_by_gender = sales.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by=['Amount'],ascending=False)
sns.barplot(x = 'Gender', y= 'Amount' , data= sales_by_gender )
```

*As the above graph we see that most buyers are females with most purchasing power over males*

2. Which age group people are more enthusiasts for shopping?

```python
ax = sns.countplot(x ='Age Group',  hue= 'Gender' , data = sales)
```

```python
Amt_agegrp = sales.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by=['Amount'],ascending=False)
sns.barplot(x ='Age Group', y= 'Amount' , data = Amt_agegrp)
```

```python
sales.info()
```

3. list the top 5 states with high numbers of Orders?

```python
Order_by_State = sales.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by=['Orders'],ascending=False).head(5)
sns.set(rc={'figure.figsize':(16,8)})
sns.barplot(x = 'State', y ='Orders', data=Order_by_State )
```

```python
#total sales from state
Sales_by_State = sales.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by=['Amount'],ascending=False)
sns.set(rc={'figure.figsize':(25,10)})
sns.barplot(x = 'State', y ='Amount', data= Sales_by_State )
```

*As above the graph we can see the most orders are from Uttar Pradesh, Maharashtra and karnataka*

4. Marital_Status

```python
ax = sns.countplot (x='Marital_Status',data=sales,palette='husl')
for bars in ax.containers:
   ax.bar_label(bars)
sns.set(rc={'figure.figsize':(5,8)})
plt.show()
```

```python
marital_stats = sales.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by=['Amount'],ascending=False)
sns.set(rc={'figure.figsize':(5,8)})
sns.barplot(x = 'Marital_Status', y ='Amount' , data=marital_stats, hue = 'Gender'  )
```

*As above graph we can see that females are most buyers are married(Women) and has high purchase power*

4. Occuption 

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot (x='Occupation',data=sales,palette='husl')
for bars in ax.containers:
   ax.bar_label(bars)
plt.show()
```

```python
Occupation = sales.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by=['Amount'],ascending=False)
sns.barplot(x = 'Occupation',y = 'Amount',data=Occupation)
```

*As in graph we can see the most buyers are from IT Sector,Healthcare & Aviation* 

5. Product_Category

```python
sns.set(rc={'figure.figsize':(30,12)})
ax = sns.countplot(x= 'Product_Category', data= sales,palette='husl')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales.groupby('Product_Category')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```

```python
#total sales by product_category
Total_sales_products = sales.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by=['Amount'],ascending=False)
sns.barplot(x = 'Product_Category',y = 'Amount',data=Total_sales_products)
sns.set(rc={'figure.figsize':(20,5)})

```

*As we can see that the most selling product_category was food,footware & Electronics category*
